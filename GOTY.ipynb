```python
#Criar arquivo PDF
import csv
from fpdf import FPDF

# Caminho do seu arquivo CSV
file_path = r'C:\Users\lucas\Downloads\GOTY\game_of_the_year.csv'

# Criar o objeto PDF
pdf = FPDF()
pdf.add_page()

# Definir o título do PDF
pdf.set_font("Arial", size=12)
pdf.cell(200, 10, txt="Relatório de Jogos Vencedores do Game of the Year", ln=True, align='C')

# Definir a fonte para os dados
pdf.set_font("Arial", size=10)

# Abrir e ler o arquivo CSV
with open(file_path, mode='r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    # Exibir os dados no PDF
    for game in reader:
        line = f"Ano: {game.get('Ano')}, Nome: {game.get('Nome')}, Nota: {game.get('Pontuação Crítica (Metacritic)')}, Popularidade: {game.get('Popularidade (jogadores)')}"
        pdf.cell(200, 10, txt=line, ln=True)

# Salvar o PDF no mesmo diretório
pdf.output(r'C:\Users\lucas\Downloads\game_of_the_year_report.pdf')
```

```python
#Criar arquivo TXT
import csv

# Caminho do seu arquivo CSV
file_path = r'C:\Users\lucas\Downloads\GOTY\game_of_the_year.csv'

# Caminho para salvar o arquivo de texto
output_path = r'C:\Users\lucas\Downloads\game_of_the_year_report.txt'

# Abrir o arquivo CSV e criar o arquivo de texto
with open(file_path, mode='r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    
    # Abrir o arquivo TXT em modo de escrita
    with open(output_path, mode='w', encoding='utf-8') as txt_file:
        # Escrever um título no arquivo de texto
        txt_file.write("Relatório de Jogos Vencedores do Game of the Year\n\n")
        
        # Escrever os dados no arquivo de texto
        for game in reader:
            line = f"Ano: {game.get('Ano')}, Nome: {game.get('Nome')}, Nota: {game.get('Pontuação Crítica (Metacritic)')}, Popularidade: {game.get('Popularidade (jogadores)')}\n"
            txt_file.write(line)

# Imprimir uma mensagem confirmando que o arquivo foi gerado
print(f"Arquivo de texto gerado em: {output_path}")
```

```python
#Criação do gráfico Jogos x Pontuação crítica
import pandas as pd
import matplotlib.pyplot as plt

try:
    # Carregar arquivo CSV
    file_path = r"C:\Users\lucas\Downloads\GOTY\game_of_the_year.csv"
    data = pd.read_csv(file_path)
    print("Arquivo carregado com sucesso!")

    # Converter e ordenar os dados
    data['Pontuação Crítica (Metacritic)'] = pd.to_numeric(data['Pontuação Crítica (Metacritic)'], errors='coerce')
    top_10_games = data.nlargest(10, 'Pontuação Crítica (Metacritic)')
    print("Top 10 jogos selecionados:\n", top_10_games[['Nome ', 'Pontuação Crítica (Metacritic)']])

    # Criar gráfico de barras
    plt.figure(figsize=(12, 6))
    plt.barh(top_10_games['Nome '], top_10_games['Pontuação Crítica (Metacritic)'], color='skyblue')
    plt.xlabel('Pontuação Crítica (Metacritic)', fontsize=12)
    plt.ylabel('Jogos', fontsize=12)
    plt.title('Top 10 Jogos com Maiores Pontuações Críticas', fontsize=14)
    plt.gca().invert_yaxis()
    plt.tight_layout()

    # Exibir o gráfico
    print("Exibindo gráfico...")
    plt.show()
except Exception as e:
    print(f"Ocorreu um erro: {e}")
```

```python
#Criação da tabela Pontuação x Data
import pandas as pd

# Carregar os dados do CSV
data = pd.read_csv('C:\\Users\\lucas\\Downloads\\GOTY\\game_of_the_year.csv')

# Agora você pode usar o groupby
average_score_per_year = data.groupby('Ano')['Pontuação Crítica (Metacritic)'].mean().reset_index()

# Criar gráfico de linhas
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(average_score_per_year['Ano'], average_score_per_year['Pontuação Crítica (Metacritic)'], marker='o', color='b', linestyle='-')
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Pontuação Média (Metacritic)', fontsize=12)
plt.title('Evolução da Pontuação Crítica ao Longo dos Anos', fontsize=14)
plt.grid(visible=True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
#Criação da tabela de Desenvolvedoras 
import matplotlib.pyplot as plt

# Contar a quantidade de vezes que cada desenvolvedora aparece
developer_counts = data['Desenvolvedora'].value_counts()

# Configurar o gráfico
plt.figure(figsize=(10, 6))
developer_counts.plot(kind='bar', color='skyblue')

# Personalizar o gráfico
plt.title('Quantidade de jogos por desenvolvedora', fontsize=14)
plt.xlabel('Desenvolvedora', fontsize=12)
plt.ylabel('Quantidade de jogos', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

```python
#Tabela Popularidade x Jogos
import pandas as pd
import matplotlib.pyplot as plt

# Caminho do arquivo CSV
caminho_arquivo = r"C:\Users\lucas\Downloads\GOTY\jogos.csv"

# Lendo o arquivo CSV
dados = pd.read_csv(caminho_arquivo)

# Selecionando as colunas "Nome" e "Popularidade (jogadores)"
jogos = dados["Nome"]
popularidade = dados["Popularidade (jogadores)"]

# Criando o gráfico
plt.figure(figsize=(12, 6))
plt.bar(jogos, popularidade, color="skyblue")
plt.xlabel("Jogo", fontsize=12)
plt.ylabel("Popularidade (Jogadores)", fontsize=12)
plt.title("Popularidade por Jogo", fontsize=14)
plt.xticks(rotation=90, fontsize=10)
plt.tight_layout()

# Exibindo o gráfico
plt.show()
```

```python
#Tabela da quantidade dos gêneros
import pandas as pd
import matplotlib.pyplot as plt

# Caminho do arquivo CSV
caminho_arquivo = r"C:\Users\lucas\Downloads\GOTY\jogos.csv"

# Lendo o arquivo CSV
dados = pd.read_csv(caminho_arquivo)

# Contando a frequência de cada gênero
# Dividindo os gêneros compostos
generos_expandido = dados["Gênero"].str.split(" / ").explode()
frequencia_generos = generos_expandido.value_counts()

# Criando o gráfico
plt.figure(figsize=(12, 6))
frequencia_generos.plot(kind="bar", color="orange", alpha=0.8)

# Ajustando o título e rótulos dos eixos
plt.xlabel("Gênero", fontsize=12)
plt.ylabel("Frequência", fontsize=12)
plt.title("Frequência dos Gêneros nos Jogos", fontsize=14)

# Ajustando a rotação e o espaçamento dos rótulos no eixo X
plt.xticks(rotation=45, ha="right", fontsize=10)  # 'ha' ajusta a posição horizontal
plt.tight_layout()  # Ajuste automático do layout para evitar sobreposição

# Exibindo o gráfico
plt.show()
```
